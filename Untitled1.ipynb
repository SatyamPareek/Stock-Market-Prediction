```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df_train=pd.read_csv('WIPRO_train.csv')
df_test=pd.read_csv("WIPRO_test.csv")
df_train.head()
```

```python
df_train=df_train.dropna()
df_test=df_test.dropna()
```

```python
train_dates=pd.to_datetime(df_train['Date'])
test_dates=pd.to_datetime(df_test['Date'])
```

```python
features=list(df_train)[1:8]
features
```

```python
df_for_training=df_train[features].astype(float)
```

```python
print(df_for_training)
```

```python
df_for_testing=df_test[features].astype(float)
```

```python
df_for_testing
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_for_training_scaled=scaler.fit_transform(df_for_training)
```

```python
trainX = []
trainY = []

n_past = 30  # Number of past days we want to use to predict the future.

#Reformat input data into a shape: (n_samples x timesteps x n_features)
for i in range(n_past, len(df_for_training_scaled)):
    trainX.append(df_for_training_scaled[i - n_past:i])
    trainY.append(df_for_training_scaled[i, 0])

trainX, trainY = np.array(trainX), np.array(trainY)

print('trainX shape == {}.'.format(trainX.shape))
print('trainY shape == {}.'.format(trainY.shape))
```

```python
from keras.models import Sequential
from keras.layers import Dense , Dropout
from keras.layers import LSTM
```

```python
model = Sequential()
#first layer
model.add(LSTM(units=64, activation='tanh', input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences=True))
model.add(Dropout(0.2))

#second layer
model.add(LSTM(units=60,activation='tanh',return_sequences=True))
model.add(Dropout(0.3))


#third layer
model.add(LSTM(units=80,activation='tanh',return_sequences=True))
model.add(Dropout(0.4))



#fourth layer
model.add(LSTM(units=100,activation='tanh'))
model.add(Dropout(0.5))


#last layer
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
# fit the model
history = model.fit(trainX, trainY, epochs=100, batch_size=32, validation_split=0.1, verbose=1)

plt.plot(history.history['loss'], label='Training loss')
plt.plot(history.history['val_loss'], label='Validation loss')
plt.legend()
```

```python
model.save('stock_wipro.h5')
```
